```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3])
b = np.array([1,2,3,4])
a,b = np.meshgrid(a, b)
print(a,b,sep='\n\n')
```

```python
fig = plt.figure(figsize=(20,8))
axes = fig.gca(projection='3d')
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a+b)
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))
axes = fig.gca(projection='3d')
axes.plot_surface(a,b,a+b,cmap='rainbow')
plt.show()
```

```python
x = np.arange(-1,1,0.02)
y = x
x,y = np.meshgrid(x, y)
```

```python
fig = plt.figure(figsize=(10,8))
axes = fig.gca(projection='3d')
axes.plot_surface(x,y,x**2+y**2,cmap='rainbow')
plt.show()
```

# Contour Plot

```python
fig = plt.figure(figsize=(10,8))
axes = fig.gca(projection='3d')
axes.contour(x,y,x**2+y**2,cmap='rainbow')
plt.show()
```
