# Gradient Descent (in general)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.arange(10)
Y = (X-5)**2
X,Y
```

```python
plt.style.use('seaborn')
```

```python
plt.plot(X,Y)
plt.xlabel('X')
plt.ylabel('F(X');
```

```python
#linear gradient to find local minima 

x = 0
lr = 0.1
error = []

plt.plot(X,Y)
plt.title('Linear gradient')
for i in range(50):
    grad = 2*(x-5)
    x -= lr*grad
    y = (x-5)**2
    error.append(y)
    plt.scatter(x,y)
plt.show()    
plt.plot(error)
plt.title('Error');
```
